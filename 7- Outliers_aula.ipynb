É criado um dado fictício de renda, e é adicionado um outlier, um bilionário e é mostrado o histograma, porém o que você deseja é saber a renda média de um americano comum.

```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
incomes = np.append(incomes, [1000000000])

import matplotlib.pyplot as plt
plt.hist(incomes, 50)
plt.show()
```

É mostrado a média, onde deveria ser algo em torno de 27000 se não houvesse um outlier.

```python
incomes.mean()
```

Feita uma função para retirar todo valor que seja 2 vezes maior ou menor que o desvio padrão, e é feito o histograma novamente.

```python
def reject_outliers(data):
    u = np.median(data)
    s = np.std(data)
    filtered = [e for e in data if (u - 2 * s < e < u + 2 * s)]
    return filtered

filtered = reject_outliers(incomes)

plt.hist(filtered, 50)
plt.show()
```

Novamente a média atualizada sem o outlier.

```python
np.mean(filtered)
```
